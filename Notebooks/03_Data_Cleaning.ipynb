## Data Cleaning & Preprocessing

### Objectives
- Load dataset
- Handle missing values
- Remove duplicates
- Remove cancelled transactions
- Remove invalid values
- Save cleaned dataset



```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)

print("Libraries Imported Successfully")
```

# load the data

```python
df = pd.read_excel("../data/raw/Online Retail.xlsx")
```

# Backup Copy

```python
df_clean = df.copy()
```

# Check Missing Values Again

```python
df_clean.isnull().sum()
```

# Missing Value Percentage

```python
missing = (df_clean.isnull().sum()/len(df_clean))*100

missing.sort_values(ascending=False)
```

# Removed Rows with Missing CustomerID

```python
df_clean = df_clean.dropna(subset=["CustomerID"])
```

```python
df_clean = df_clean.dropna(subset=["CustomerID"])
df_clean.isnull().sum()
```

# Removed Missing Description

```python
df_clean = df_clean.dropna(subset=["Description"])
```

# Dataset Size

```python
print(df_clean.shape)
```

# Checked Duplicate Rows

```python
df_clean.duplicated().sum()
```

# clean duplicate

```python
df_clean = df_clean.drop_duplicates()
```

```python
df_clean.duplicated().sum()
```

```python
df_clean[df_clean["InvoiceNo"].astype(str).str.startswith("C")].head()
```

# Remove Cancelled Orders

```python
df_clean["InvoiceNo"].astype(str).str.startswith("C").sum()
```

```python
df_clean = df_clean[
    ~df_clean["InvoiceNo"].astype(str).str.startswith("C")
]
```

# Remove Negative Quantities

```python
df_clean = df_clean[df_clean["Quantity"] > 0]
```

# Remove Zero or Negative Prices

```python
df_clean = df_clean[df_clean["UnitPrice"] > 0]
```

# Convert CustomerID to Integer

```python
df_clean["CustomerID"].dtype
```

```python
df_clean["CustomerID"] = df_clean["CustomerID"].astype(int)
```

# Convert InvoiceDate

```python
df_clean["InvoiceDate"].dtype
```

```python
df_clean["InvoiceDate"] = pd.to_datetime(df_clean["InvoiceDate"])
df_clean["InvoiceDate"].dtype
```

# Final Dataset Info

```python
df_clean.info()
```

# Final Shape


```python
print("Rows:", df_clean.shape[0])
print("Columns:", df_clean.shape[1])
```

# Verify Missing Values

```python
df_clean.isnull().sum()
```

# Verify Duplicates

```python
df_clean.duplicated().sum()
```

# Save Clean Dataset

```python
df_clean.to_csv(
    "../data/processed/online_retail_clean.csv",
    index=False
)
```
